```julia
module BipedRecovery
using JuMP
using Gurobi
export BipedState, RecoveryProblem, setup_vars!

type BipedState
    xcom::Array{Float64,1}
    qcop::Array{Float64,1}
    qr::Array{Float64,1}
    ql::Array{Float64,1}
    contact::Array{Bool,1}
end

type RecoveryProblem
    start::BipedState
    nsteps::Integer
    m::Model
    omega::FloatingPoint
    dt::FloatingPoint
    xcom
    qcop
    qr
    ql
    contained
    
    RecoveryProblem(start::BipedState, nsteps::Integer) = new(start, 
    nsteps, Model(solver=GurobiSolver(OutputFlag=0)))
end

function setup_vars!(prob::RecoveryProblem)
    @defVar(prob.m, xcom[1:4, 1:prob.nsteps])
    prob.xcom = xcom
    @defVar(prob.m, qcop[1:2, 1:prob.nsteps])
    prob.qcop = qcop
    @defVar(prob.m, qr[1:2, 1:prob.nsteps])
    prob.qr = qr
    @defVar(prob.m, ql[1:2, 1:prob.nsteps])
    prob.ql = ql
    @defVar(prob.m, contained[1:2, 1:prob.nsteps], Bin)
    prob.contained = contained
    
    for varname = {"xcom", "qcop", "qr", "ql"}
        var = prob.(symbol(varname))
        for m = 1:length(var) / prob.nsteps
            @addConstraint(prob.m, var[m,1] == prob.start.(symbol(varname))[m])
        end
    end
    for n = 2:prob.nsteps
        for m = 1:2
            for var = {prob.xcom, prob.qcop, prob.qr, prob.ql}
                setLower(var[m,n], prob.start.xcom[m] - 30)
                setUpper(var[m,n], prob.start.xcom[m] + 30)
            end
        end
        for m = 3:4
            setLower(prob.xcom[m,n], prob.start.xcom[m] - 10)
            setUpper(prob.xcom[m,n], prob.start.xcom[m] + 10)
        end
    end
            
    return prob
end

function add_initial_contact_constraints!(prob::RecoveryProblem)
    if prob.start.contact[1]
        @addConstraint(prob.m, r[m=1:2], prob.qr[m,1] == prob.qr[m,2])
    end
    if prob.start.contact[2]
        @addConstraint(prob.m, l[m=1:2], prob.ql[m,1] == prob.ql[m,2])
    end
end

function add_discrete_linear_dynamics!(prob::RecoveryProblem)
    A = Float64[0 0 1 0;
                0 0 0 1;
        prob.omega^2 0 0 0;
        0 prob.omega^2 0 0]
    B = [0 0;
         0 0;
        -prob.omega^2 0;
        0 -prob.omega^2]
    Ai = inv(A)
    exAdt = expm(A * prob.dt)
    
    N = prob.nsteps-1
    @defExpr(alpha[m=1:2, n=1:N], (prob.qcop[m, n+1] - prob.qcop[m, n]) / prob.dt)
    
    @defExpr(beta[m=1:2, n=1:N], prob.qcop[m,n])
    
    @defExpr(T[m=1:4, n=1:N], dot(vec((-Ai * B)[m,:]), vec(beta[:,n])) + dot(vec((-Ai * Ai * B)[m,:]), vec(alpha[:,n])))
    
    @defExpr(S[m=1:4, n=1:N], dot(vec((-Ai * B)[m,:]), vec(alpha[:,n])))
    
    @defExpr(Q[m=1:4, n=1:N], prob.xcom[m,n] + dot(vec((Ai * B)[m,:]), vec(beta[:,n])) + dot(vec((Ai * Ai * B)[m,:]), vec(alpha[:,n])))
    
    @addConstraint(prob.m, dynamics_constr[m=1:4, n=1:N], prob.xcom[m,n+1] == dot(vec(exAdt[m,:]), vec(Q[:,n])) + prob.dt * S[m,n]  + T[m,n])
    
    @addConstraint(prob.m, final_cop[m=1:2], prob.qcop[m,end] == 0.5 * (prob.qr[m,end] + prob.ql[m,end]))
end

function add_reachability!(prob::RecoveryProblem)
    max_forward_step = max(0.2, prob.start.qr[1] - prob.start.xcom[1], prob.start.ql[1] - prob.start.xcom[1])
    max_backward_step = min(-0.15, prob.start.qr[1] - prob.start.xcom[1], prob.start.ql[1] - prob.start.xcom[1])
    r_limits = [max_backward_step max_forward_step;
                -0.25 -0.1]
    
    l_limits = [max_backward_step max_forward_step;
                0.1 0.25]
    N = prob.nsteps
    @addConstraint(prob.m, r_bounds[m=1:2, n=2:N], r_limits[m,1] <= prob.qr[m,n] - prob.xcom[m,n])
    @addConstraint(prob.m, r_bounds[m=1:2, n=2:N], prob.qr[m,n] - prob.xcom[m,n] <= r_limits[m,2])
    
    @addConstraint(prob.m, l_bounds[m=1:2, n=2:N], l_limits[m,1] <= prob.ql[m,n] - prob.xcom[m,n])
    @addConstraint(prob.m, l_bounds[m=1:2, n=2:N], prob.ql[m,n] - prob.xcom[m,n] <= l_limits[m,2])
    
    @addConstraint(prob.m, l_minus_r[m=2, n=2:N], prob.ql[m,n] - prob.qr[m,n] <= 0.45)
    @addConstraint(prob.m, l_minus_r[m=2, n=2:N], 0.16 <= prob.ql[m,n] - prob.qr[m,n])
    
end

function abs(e::AffExpr)
    m = e.vars[1].m
    @defVar(m, aux >= 0)
    @addConstraint(m, aux >= e)
    @addConstraint(m, aux >= -e)
    return aux
end

function abs(v::Variable)
  @defVar(v.m, aux >= 0)
  @addConstraint(v.m, aux >= v)
  @addConstraint(v.m, aux >= -v)
  return aux
end

function prevent_motion!(prob::RecoveryProblem, js, dj, stance_upper_bound)
    @addConstraint(prob.m, fixed[m=1:2, j=js], abs(prob.qr[m,j+dj+1] - prob.qr[m,j+dj]) <= stance_upper_bound * (1-prob.contained[1,j]))
    @addConstraint(prob.m, fixed[m=1:2, j=js], abs(prob.ql[m,j+dj+1] - prob.ql[m,j+dj]) <= stance_upper_bound * (1-prob.contained[2,j]))
end
    

function add_contact_constraints!(prob::RecoveryProblem)
    foot_bounds = [-0.05 0.05;
                   -0.02 0.02]
    stance_upper_bound = 2.0
    
    @addConstraint(prob.m, contained[n=1:prob.nsteps], sum(prob.contained[:,n]) == 1)
    
    @addConstraint(prob.m, motion[m=1:2, j=2:prob.nsteps-1], foot_bounds[m,1] - stance_upper_bound * (1-prob.contained[1,j]) <= prob.qcop[m,j] - prob.qr[m,j])
    @addConstraint(prob.m, motion[m=1:2, j=2:prob.nsteps-1], prob.qcop[m,j] - prob.qr[m,j] <= foot_bounds[m,2] + stance_upper_bound * (1-prob.contained[1,j]))
    @addConstraint(prob.m, motion[m=1:2, j=2:prob.nsteps-1], foot_bounds[m,1] - stance_upper_bound * (1-prob.contained[2,j]) <= prob.qcop[m,j] - prob.ql[m,j])
    @addConstraint(prob.m, motion[m=1:2, j=2:prob.nsteps-1], prob.qcop[m,j] - prob.ql[m,j] <= foot_bounds[m,2] + stance_upper_bound * (1-prob.contained[2,j]))
    
    prevent_motion!(prob, 1:prob.nsteps-1, 0, stance_upper_bound)
    prevent_motion!(prob, 2:prob.nsteps-1, -1, stance_upper_bound)
    prevent_motion!(prob, 3:prob.nsteps-1, -2, stance_upper_bound)
    prevent_motion!(prob, 1:prob.nsteps-2, 1, stance_upper_bound)
end

function get_capture_point_objective(prob::RecoveryProblem)
    return abs(prob.xcom[1,end] + 1/prob.omega * prob.xcom[3,end] - 0.5 * (prob.ql[1,end] + prob.qr[1,end])) + abs(prob.xcom[2,end] + 1/prob.omega * prob.xcom[4,end] - 0.5 * (prob.ql[2,end] + prob.qr[2,end]))
end

function get_foot_motion_objective(prob::RecoveryProblem)
    foot_motion_weight = 0.01
    @defExpr(motion_obj[j=1:prob.nsteps-1], foot_motion_weight * (abs(prob.qr[1,j] - prob.qr[1,j+1]) + abs(prob.qr[2,j] - prob.qr[2,j+1]) + abs(prob.ql[1,j] - prob.ql[1,j+1]) + abs(prob.ql[2,j] - prob.ql[2,j+1])))
    return sum(motion_obj)
end

function get_final_posture_objective(prob::RecoveryProblem)
    nom_stance_width = 0.26
    nom_stance_r = [0, -nom_stance_width/2]
    nom_stance_l = [0, nom_stance_width/2]
    final_posture_weight = 0.01
    @defExpr(posture_obj[m=1:2], final_posture_weight * (abs(prob.qr[m,end] - prob.xcom[m,end] - nom_stance_r[m]) + abs(prob.ql[m,end] - prob.xcom[m,end] - nom_stance_l[m])))
    return sum(posture_obj)
end

function add_foot_velocity_limits!(prob::RecoveryProblem)
    max_foot_velocity = 2
    @addConstraint(prob.m, vel[m=1:2, j=1:prob.nsteps-1], prob.qr[m,j+1] - prob.qr[m,j] <= max_foot_velocity * prob.dt)
    @addConstraint(prob.m, vel[m=1:2, j=1:prob.nsteps-1], prob.ql[m,j+1] - prob.ql[m,j] <= max_foot_velocity * prob.dt)
end

function add_final_foot_velocity(prob::RecoveryProblem)
    @addConstraint(prob.m, vel[m=1:2], prob.qr[m,end-1] == prob.qr[m,end])
    @addConstraint(prob.m, vel[m=1:2], prob.ql[m,end-1] == prob.ql[m,end])
end

function setupAndSolve(prob::RecoveryProblem)
    setup_vars!(prob)
    add_initial_contact_constraints!(prob)
    add_discrete_linear_dynamics!(prob)
    add_reachability!(prob)
    add_contact_constraints!(prob)
    add_foot_velocity_limits!(prob)
    add_final_foot_velocity(prob)
    
    @setObjective(prob.m, Min, get_capture_point_objective(prob) + get_foot_motion_objective(prob) + get_final_posture_objective(prob))
    
    solve(prob.m)
    getObjectiveValue(prob.m)
end


end
```

```julia
import BipedRecovery
```

```julia
nsteps = 15
dt = 0.05
omega = sqrt(9.81 / 0.84)
start = BipedRecovery.BipedState(Float64[0, 0, 0.4, 0],
Float64[0,0],
Float64[0,-0.13],
Float64[0,0.13],
Bool[1,1])

@time begin
prob = BipedRecovery.RecoveryProblem(start, nsteps)
prob.omega = omega
prob.dt = dt
BipedRecovery.setupAndSolve(prob)
end

@time begin
prob = BipedRecovery.RecoveryProblem(start, nsteps)
prob.omega = omega
prob.dt = dt
BipedRecovery.setupAndSolve(prob)
end
```

```julia
xcom = getValue(prob.xcom)[:,:]
qcop = getValue(prob.qcop)[:,:]
qr = getValue(prob.qr)[:,:]
ql = getValue(prob.ql)[:,:]
contained = getValue(prob.contained)[:,:]

N = size(xcom)[2]
figure()
hold(true)
plot(vec(xcom[1,:]), vec(xcom[2,:]), "-")
plot(vec(qcop[1,:]), vec(qcop[2,:]), "-", color=[.8, .5, .1])
for j = 1:N
    color = (j/N) * [0,0,1] + ((N-j)/N) * [0,1,0]
    plot(xcom[1,j], xcom[2,j], "o", color=color)
    plot(qcop[1,j], qcop[2,j], "o", color=color)
end
```
